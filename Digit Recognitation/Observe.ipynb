```python
import json
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
f = open('Report_network2.json', 'r')
data = json.load(f)
f.close
```

```python
data = pd.DataFrame(data)
```

```python
data.head()
```

```python
fig = plt.figure()
a = fig.add_subplot(111)
x = np.arange(0,30)

y = [(i*100)/10000 for i in data['train_acc']]
a.plot(x,y,color='#FFA933',label="Accuracy on the training data")

y = [(i*100)/1000 for i in data['eval_acc']]
a.plot(x,y,color='#2A6EA6',label="Accuracy on the test data")
plt.ylim(0,100)
plt.xlim(1,30)
plt.legend(loc="lower right")
plt.show()
```

```python
a = fig.add_subplot(111)
x = np.arange(0,30)

y = data['train_cost']
plt.plot(x,y,color='r',label="Cost on the training data")

y = data['eval_cost']
plt.plot(x,y,color='b',label="Cost on the testing data")
plt.legend(loc="upper right")
plt.show()
```
