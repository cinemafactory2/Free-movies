```python
import numpy as np
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup as soup
import time, math
```

```python
def wait(driver):
        movie_elements = driver.find_elements_by_class_name('viewAllGrid')
        
        n = len(movie_elements)
        for i in range(0,n,10):
            driver.execute_script('arguments[0].scrollIntoView();',movie_elements[i])
            time.sleep(1)
            percant = math.trunc((i/n)*100)
            print('{0}% ->'.format(percant), end = ' ')
        print('100%.')
        
        page = driver.execute_script("return document.documentElement.outerHTML")
        page_html = soup(page, 'html.parser')
        movies = page_html.findAll(class_ =  'viewAllGrid')
        
        return movies
```

```python
def loadFully(driver, source):
    try:
        driver.get(source)
    except:
        return 0
    
    end = driver.find_element_by_class_name('footerContainer')
    check = 0
    previous_len = 0
    movie = []
    while True:
        driver.execute_script('arguments[0].scrollIntoView();',end)
        #time.sleep(1)
        page = driver.execute_script("return document.documentElement.outerHTML")
        page_html = soup(page, 'html.parser')
        movies = page_html.findAll(class_ =  'viewAllGrid')
        curr_len = len(movies)
        if curr_len != previous_len:
            check = 0
            print('LOADING...', curr_len)
        if curr_len == previous_len:
            check += 1
            if check == 5:
                print('LOADING...', curr_len)
                break
        previous_len = curr_len
    
    movies = wait(driver) #this wait is for load photos
    print('DONE\nTotal Movies Captured : ', curr_len)
    return movies
```

```python
def isFree(movie):
    isPremium = movie.find(class_ = 'cardPremiumContent')
    if isPremium:
        return False
    else:
        return True
```

```python
def getMovieDetail(movie):
    try:
        movie_link = movie.div['data-minutelyurl']
    except:
        movie_link = None

    content = movie.find(class_ = 'content')
    try:
        movie_title = content['data-minutelytitle']
    except:
        movie_title = None
    
    try:
        movie_img_src = content.img['src']
    except:
        movie_img_src = None
        
    return movie_title, movie_link, movie_img_src
```

```python
def visitSource(driver, Source, Category_):
    Movie = []
    Link = []
    Image = []
    Category = []

    i = 1
    n = len(Source)
    x = 0
    for source in Source:
    #for y in range(3):
        free = 0
        print('Running Source :',i)
        #source = Source[y]
        category = Category_[x]
        x += 1
        movies = loadFully(driver, source)
        if movies == 0:
            print('Source {0} can not be Loaded.', i)
            continue
        
        for movie in movies:
            movie_title, movie_link, movie_img_src = getMovieDetail(movie)
    
            if isFree(movie):
                Movie.append(movie_title)
                Link.append(movie_link)
                Image.append(movie_img_src)
                Category.append(category)
                free += 1

        print('Free Movies :',free)
        print('Source Complete : {0}/{1}.\n'.format(i,n))
        i += 1
    
    return Movie, Link, Image, Category
```

```python
option = webdriver.ChromeOptions()
option.add_argument('headless')
driver = webdriver.Chrome(options = option)

CategoryData = pd.read_csv('CategoryZEE5.csv')
Source = list(CategoryData['Source'])
Category_ = list(CategoryData['Category'])
    
Movie, Link, Image, Category = visitSource(driver, Source, Category_)

driver.quit()
    
MovieData = pd.DataFrame({
    'Movie' : Movie,
    'Link'  : Link,
    'Image_src' : Image,
    'Category' : Category
}) 
```
